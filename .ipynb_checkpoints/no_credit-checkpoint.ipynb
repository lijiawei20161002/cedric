```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ddos_gym/ddos_gym/envs/data/attack.csv')
display(df)
```

```python
attack_frequency = {}

for cl in df['dst_cc']:
    cl = cl.split('[')[1].split(']')[0].split(",")
    for c in cl:
        c = c.split("\'")[1].split("\'")[0]
        if c not in attack_frequency:
            attack_frequency[c] = 1
        else:
            attack_frequency[c] += 1

plt.xticks(rotation=70)
plt.bar(list(attack_frequency.keys())[:10], list(attack_frequency.values())[:10])
```
